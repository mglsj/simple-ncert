```python
import pandas as pd
import matplotlib.pyplot as plt

# read the CSV file with specified columns
# usecols parameter to extract only two required columns
data = pd.read_csv(
    "Min_Max_Seasonal_IMD_2017.csv",
    usecols=["ANNUAL - MIN", "ANNUAL - MAX"],
)
df = pd.DataFrame(data)
# plot histogram for 'ANNUAL - MIN'
df.plot(
    kind="hist",
    y="ANNUAL - MIN",
    title="Annual Minimum Temperature (1901-2017)",
)
plt.xlabel("Temperature")
plt.ylabel("Number of times")

plt.savefig("fig_4.11.webp", bbox_inches="tight")

# plot histogram for both 'ANNUAL - MIN' and 'ANNUAL - MAX'
df.plot(
    kind="hist",
    title="Annual Min and Max Temperature (1901-2017)",
    color=["blue", "red"],
)
plt.xlabel("Temperature")
plt.ylabel("Number of times")
# plt.show()

plt.savefig("fig_4.12.webp", bbox_inches="tight")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(
    "Min_Max_Seasonal_IMD_2017.csv",
    usecols=["ANNUAL - MIN"],
)
df = pd.DataFrame(data)
# convert the 'ANNUAL - MIN' column into a numpy 1D array
minarray = np.array([df["ANNUAL - MIN"]])
# Extract y (frequency) and edges (bins)
y, edges = np.histogram(minarray)
# calculate the midpoint for each bar on the histogram
mid = 0.5 * (edges[1:] + edges[:-1])
df.plot(kind="hist", y="ANNUAL - MIN")
plt.plot(mid, y, "-^")
plt.title("Annual Min Temperature plot(1901 - 2017)")
plt.xlabel("Temperature")
# plt.show()

plt.savefig("fig_4.13.webp", bbox_inches="tight")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(
    {
        "GeoArea": [83743, 78438, 22327, 22429, 21081, 16579, 10486],
        "ForestCover": [67353, 27692, 17280, 17321, 19240, 13464, 8073],
    },
    index=[
        "Arunachal Pradesh",
        "Assam",
        "Manipur",
        "Meghalaya",
        "Mizoram",
        "Nagaland",
        "Tripura",
    ],
)
exp = [0.1, 0, 0, 0, 0.2, 0, 0]
# explode the first wedge to .1 level and fifth to level 2.
c = ["r", "g", "m", "c", "brown", "pink", "purple"]
# change the color of each wedge
df.plot(
    kind="pie",
    y="ForestCover",
    title="Forest cover of NorthEastern states",
    legend=False,
    explode=exp,
    autopct="%.2f",
    colors=c,
)
# plt.show()

plt.savefig("fig_4.22.webp", bbox_inches="tight")
```
